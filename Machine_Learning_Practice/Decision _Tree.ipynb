```python
df= pd.get_dummies(df,columns=['color'])
```

```python

for i,row in df.iterrows():
    #print(i == 'Ghoul')
    #print(i)
    #print(row['type'])
    type_int=0
    if row['type'] == 'Ghoul':
        type_int=int(1)
    elif row['type'] == 'Goblin':
        type_int=int(2)
    else:
        type_int=int(3)
    df.at[i,'type_int'] = int(type_int)
    #break
```

```python
df['type_int'] = pd.to_numeric(df['type_int'],downcast='integer')
```

```python
df.head()
```

```python
X_train = df
```

```python
dt= tree.DecisionTreeClassifier(random_state=0)
```

```python
X_train.columns
```

```python
X_train =df[[ 'bone_length', 'rotting_flesh', 'hair_length', 'has_soul',
       'color_black', 'color_blood', 'color_blue', 'color_clear',
       'color_green', 'color_white']]
```

```python
Y_train = df['type_int']
```

```python
dt.fit(X_train,Y_train)
```

```python
#Pydotplus pakage is used when we want to use 3rd party pakages. 
#1. create IO obj for writing and reading 3rd party object
objstringIO = io.StringIO()
#write the graphviz object to objstringIO(object)
tree.export_graphviz(dt,out_file=objstringIO, feature_names = X_train.columns,filled=True, rounded=True)
#Use out_file = objStringIO to getvalues()
#will read objstringIO object content line by line
file1 = pydotplus.graph_from_dot_data(objstringIO.getvalue())
#os.chdir('C:/Users/kumarijy/Documents/Learning/gyanVriksh/Titanic')
#write read content to pdf
```

```python
file1.write_pdf("DecissionTree_GGG.pdf")
```

```python
from sklearn.model_selection import cross_val_score
clf = tree.DecisionTreeClassifier(random_state=0)
cross_val_score(clf,X_train,Y_train)
```
